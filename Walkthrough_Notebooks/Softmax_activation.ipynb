# Softmax activation

```python
from math import exp
```

exp is just shorthand for exponential (evaluates argument raised to the power of e)

#### 

- y1 hat: output to predict hotdog
- y2 hat: output to predict burger
- y3 hat: output to predict pizza

```python
z = [-1, 1, 5] #value of the activation y1, y2, y3 respectively in the output layer (NOT the y-hat values)
```

```python
exp(z[0])
```

```python
exp(z[1])
```

```python
exp(z[2])
```

```python
#sum up the exponentials
total = 0
for i in range(len(z)):
    total += exp(z[i])

print(total)
```

```python
#Divide each value of exp(z[i]) by the total sum of all exponentials to get the corresponding y-hat value
print("y1-hat: ", exp(z[0])/total)
print("y2-hat: ", exp(z[1])/total)
print("y3-hat: ", exp(z[2])/total)
```

From this data, our model predicts:
- 0.2% chance of being a hot dog
- 1.8% chance of being a burger
- 97% chance of being a pizza

#### Softmax provides a likelihood for each category it tries to classify an input as (provides confidence for each)
